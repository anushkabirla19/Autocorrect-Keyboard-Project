```python
import re
from collections import Counter
```

```python
words = []
with open('/content/drive/MyDrive/boon.txt', 'r') as f:  #The 'r' means that the file should be opened in read mode.
    file_name_data = f.read()
    file_name_data=file_name_data.lower()  #converting every word to lower case.
    words = re.findall('\w+',file_name_data)  #re.findall function returns all the words from the file in the list.
# This is our vocabulary
V = set(words)  #creating a set of all the words present in the file.
print(f"The first ten words in the text are: \n{words[0:10]}")
print(f"There are {len(V)} unique words in the vocabulary.")
```

```python
word_freq_dict = {}
word_freq_dict = Counter(words)  #each word will be a key and the value assigned to each word will be the number of times it has appeared in the list of words
print(word_freq_dict.most_common()[0:10])  #top ten words with highest frequncy.
```

```python
probs = {}  #probs is the required dictionary that will contain each word w and the respective probability P(w).
Total = sum(word_freq_dict.values())
for k in word_freq_dict.keys():
    probs[k] = word_freq_dict[k]/Total
```

```python
def my_autocorrect(input_word):
    input_word = input_word.lower()
    if input_word in V:
        return('Your word seems to be correct')
    else:
        similarities = [1-(textdistance.Jaccard(qval=2).distance(v,input_word)) for v in word_freq_dict.keys()]
        df = pd.DataFrame.from_dict(probs, orient='index').reset_index()
        df = df.rename(columns={'index':'Word', 0:'Prob'})
        df['Similarity'] = similarities
        output = df.sort_values(['Similarity', 'Prob'], ascending=False).head()  #return the 5 most similar words ordered by similarity and probability
        return(output)
```

```python
my_autocorrect('nevrtless')
```
